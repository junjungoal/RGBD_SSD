```python
import os, sys
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
rgb_precision_f = open('rgb_precision.csv', 'r')
rgb_recall_f = open('rgb_recall.csv', 'r')

rgbd_precision_f = open('rgbd-3_precision.csv', 'r')
rgbd_recall_f = open('rgbd-3_recall.csv', 'r')
rgb_pre_reader = csv.reader(rgb_precision_f)
rgb_recall_reader = csv.reader(rgb_recall_f)

rgbd_pre_reader = csv.reader(rgbd_precision_f)
rgbd_recall_reader = csv.reader(rgbd_recall_f)
```

```python
rgb_precision = []
for pre in rgb_pre_reader:
    rgb_precision.append(pre)
    
rgb_recall = []
for pre in rgb_recall_reader:
    rgb_recall.append(pre)
    
rgbd_precision = []
for pre in rgbd_pre_reader:
    rgbd_precision.append(pre)
    
rgbd_recall = []
for pre in rgbd_recall_reader:
    rgbd_recall.append(pre)
```

```python
plt.figure
plt.plot(rgb_recall[4], rgb_precision[4], label='RGB')
plt.plot(rgbd_recall[4], rgbd_precision[4], label='RGBD')
plt.axis([0, 1.0, 0, 1.0])
plt.legend(bbox_to_anchor=(0.665, 0.9), loc=2, borderaxespad=0.)
plt.title('chair')
plt.xlabel('Recall')
plt.ylabel('Precision')
#plt.savefig('./PRC_chair.png')
plt.show()
```
